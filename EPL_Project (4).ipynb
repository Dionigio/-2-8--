```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

matches_data = pd.read_csv('data/football_matches.csv')

matches_data['home_goals'] = pd.to_numeric(matches_data['home_goals'])
matches_data['away_goals'] = pd.to_numeric(matches_data['away_goals'])
matches_data['home_team_rating'] = pd.to_numeric(matches_data['home_team_rating'])
matches_data['away_team_rating'] = pd.to_numeric(matches_data['away_team_rating'])

plt.figure(figsize=(10, 6))
sns.histplot(matches_data['home_goals'] + matches_data['away_goals'], kde=True)
plt.title('    ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='home_team_rating', y='home_goals', data=matches_data)
plt.title('          ')
plt.xlabel('  ')
plt.ylabel('   ')
plt.show()\

plt.figure(figsize=(8, 6))
correlation_matrix = matches_data[['home_goals', 'away_goals', 'home_team_rating', 'away_team_rating']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('  ')
plt.show()

correlation = matches_data['home_team_rating'].corr(matches_data['home_goals'])
print(f'  : {correlation}')
```
